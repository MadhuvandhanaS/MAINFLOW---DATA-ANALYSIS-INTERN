```python
plt.scatter(data['Date'],data['Volume'])

# adding title to the plot
plt.title("Scatter Plot")

#Scattering the x and y labels
plt.xlabel('Date')
plt.ylabel('Volume')

#show the result
plt.show()
```

```python
plt.scatter(data['Date'],data['Volume'])

# adding title to the plot
plt.title("Line chart")

#Scattering the x and y labels
plt.xlabel('Date')
plt.ylabel('Volume')

#show the result
plt.show()
```

```python
#bar chart or bar plot
plt.bar(data['Date'],data['Volume'])

# adding title to the plot
plt.title("Bar chart")

#Scattering the x and y labels
plt.xlabel('Date')
plt.ylabel('Volume')

#show the result
plt.show()
```
