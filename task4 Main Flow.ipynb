```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

# Load your CSV data into a DataFrame
df = pd.read_csv('C:\\Users\\madhu_0d\\Downloads\\Telsa Stock Dataset\\TSLA-2.csv')  # Replace 'data.csv' with your actual file path

# Convert the 'Date' column to datetime format if not already done
df['Date'] = pd.to_datetime(df['Date'])

# Extract the year from the 'Date' column
df['Year'] = df['Date'].dt.year

# Group by 'Year' and calculate the mean of 'Open' prices
yearly_open = df.groupby('Year')['Open'].mean()

# Plotting the bar chart
plt.figure(figsize=(10, 6))  # Set the size of the figure
plt.bar(yearly_open.index, yearly_open.values, color='skyblue')

# Adding titles and labels
plt.title('Average Open Price by Year')
plt.xlabel('Year')
plt.ylabel('Average Open Price')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Load your CSV data into a DataFrame
df = pd.read_csv('C:\\Users\\madhu_0d\\Downloads\\Telsa Stock Dataset\\TSLA-2.csv')  # Replace 'data.csv' with your actual file path

# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Extract the year from the 'Date' column
df['Year'] = df['Date'].dt.year

# Display summary statistics
print(df.describe())

# Plotting the line chart for trends
plt.figure(figsize=(14, 7))
plt.plot(df['Date'], df['Open'], label='Open', color='blue', linestyle='--')
plt.plot(df['Date'], df['Close'], label='Close', color='red')
plt.plot(df['Date'], df['High'], label='High', color='green')
plt.plot(df['Date'], df['Low'], label='Low', color='orange')
plt.title('Trends of Open, Close, High, and Low Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()

# Plotting box plots for distribution
plt.figure(figsize=(14, 7))
sns.boxplot(data=df[['Open', 'Close', 'High', 'Low']])
plt.title('Box Plot of Open, Close, High, and Low Prices')
plt.ylabel('Price')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Load your CSV data into a DataFrame
df = pd.read_csv('C:\\Users\\madhu_0d\\Downloads\\Telsa Stock Dataset\\TSLA-2.csv')  # Replace 'data.csv' with your actual file path

# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Extract the year from the 'Date' column
df['Year'] = df['Date'].dt.year

# Plotting histograms for frequency distribution
plt.figure(figsize=(14, 10))
df['Open'].hist(bins=20, color='blue', alpha=0.7, label='Open')
df['Close'].hist(bins=20, color='red', alpha=0.7, label='Close')
df['High'].hist(bins=20, color='green', alpha=0.7, label='High')
df['Low'].hist(bins=20, color='orange', alpha=0.7, label='Low')

plt.title('Histograms of Open, Close, High, and Low Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.legend()
plt.show()


import seaborn as sns
import matplotlib.pyplot as plt

# Correlation heatmap
plt.figure(figsize=(8, 6))
correlation_matrix = df[['Open', 'Close', 'High', 'Low']].corr()

# Plotting the heatmap with seaborn
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Prices')
plt.show()
```
